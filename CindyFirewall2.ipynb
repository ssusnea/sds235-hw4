```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# load raw data
firewall_04062012 = pd.read_csv('Firewall/Firewall-04062012.csv',low_memory=False)
firewall_04072012 = pd.read_csv('Firewall/Firewall-04072012.csv',low_memory=False)
```

# Data Wrangling

```python
# filter by operation
filtered0406 = firewall_04062012[firewall_04062012['Operation'].isin(['Deny','Deny by ACL'])]
filtered0407 = firewall_04072012[firewall_04072012['Operation'].isin(['Deny','Deny by ACL'])]
```

```python
print(filtered0406['Operation'].unique())
print(filtered0407['Operation'].unique())
```

```python
print(filtered0406['Source port'].nunique())
print(filtered0407['Source port'].nunique())
```

```python
print(firewall_04062012['Source port'].nunique())
print(firewall_04072012['Source port'].nunique())
```

```python
# Count and sort for filtered0406
count_0406 = filtered0406['Source port'].value_counts().sort_values(ascending=False)
print(count_0406)

# Count and sort for filtered0407
count_0407 = filtered0407['Source port'].value_counts().sort_values(ascending=False)
print(count_0407)
```

```python
# Count the occurrences of each unique source port, sort them, and take the top 10
top_10_source_ports0406 = filtered0406['Source port'].value_counts().sort_values(ascending=False).head(10)
top_10_source_ports0407 = filtered0407['Source port'].value_counts().sort_values(ascending=False).head(10)

# Display the top 10 source ports
print(top_10_source_ports0406)
print(top_10_source_ports0407)
```

```python
# Create a list of the top 10 source ports (the index of the Series)
top_10_list0406 = top_10_source_ports0406.index.tolist()
top_10_list0407 = top_10_source_ports0407.index.tolist()

# Display the list
print(top_10_list0406)
print(top_10_list0407)
```

```python
filtered0406_top10 = filtered0406[filtered0406['Source port'].isin(top_10_list0406)]
filtered0406_top10
```

```python
filtered0407_top10 = filtered0407[filtered0407['Source port'].isin(top_10_list0407)]
filtered0407_top10
```

```python
# Count occurrences of each 'Source port' value, grouped by 'unique source port'
filtered0406_top10_count = filtered0406_top10.groupby('Source port').size().reset_index(name='count_source_port')
filtered0406_top10_count
```

```python
# bar plot for top 10 denied source port 04/05-04/06

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=filtered0406_top10_count, x='Source port', y='count_source_port')

# Customizing the plot
plt.title("Counting Firewall Records: Top 10 Source Ports Deny/Deny by ACL (04/05-04/06)")
plt.xlabel("Source Port")
plt.ylabel("Count")
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Count occurrences of each 'Source port' value, grouped by 'unique source port'
filtered0407_top10_count = filtered0407_top10.groupby('Source port').size().reset_index(name='count_source_port')
filtered0407_top10_count
```

```python
# bar plot for top 10 denied source port 04/06-04/07

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=filtered0407_top10_count, x='Source port', y='count_source_port')

# Customizing the plot
plt.title("Counting Firewall Records: Top 10 Source Ports Deny/Deny by ACL (04/06-04/07)")
plt.xlabel("Source Port")
plt.ylabel("Count")
plt.tight_layout()

# Display the plot
plt.show()
```
